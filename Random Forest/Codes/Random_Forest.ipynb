```python
from sklearn.datasets import load_digits
d = load_digits()
d
```

```python
print(list(d.keys()))
```

```python
import pandas as pd
df = pd.DataFrame(d['data'],columns = d['feature_names'])
df
```

```python
df['target'] = d['target']
print(len(df['target']))
```

```python
df
```

```python
x = df.drop('target',axis = 'columns')
y = df['target']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
print(len(x_train))
print(len(x_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

model = RandomForestClassifier()          #model = RandomorestClassifier(n_estimators = 10)
model1 = DecisionTreeClassifier()
model2 = LogisticRegression()
model3 = SVC()
model4 = KNeighborsClassifier()

model.fit(x_train,y_train)
model1.fit(x_train,y_train)
model2.fit(x_train,y_train)
model3.fit(x_train,y_train)
model4.fit(x_train,y_train)

y_pred = model.predict(x_test)
print(y_pred)
```

```python
print("Random Forest Acuraccy" ,model.score(x_test,y_test))
print("Decision Tree Acurracy" ,model1.score(x_test,y_test))
print("Logistic Regression" ,model2.score(x_test,y_test))
print("SVM" ,model3.score(x_test,y_test))
print("KNN Accuracy" ,model4.score(x_test,y_test))
```

```python
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,7))
plt.xlabel('Predicted')
plt.ylabel('Truth')

import seaborn as sns
sns.heatmap(cm,annot = True)
plt.show()
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred))
```
