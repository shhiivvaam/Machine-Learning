# **scatter**

```python
import matplotlib.pyplot as plt
import pandas as pd
#x=[10,20,30,40,50]
#y=[1,2,3,4,5]
#plt.scatter(x,y)
df=pd.read_csv('/content/IRIS.csv')

plt.scatter(df.sepal_length,df.sepal_width)
plt.show()
```

```python
x=[10,20,30,40,50]
y=[1,2,3,4,5]
plt.scatter(x,y,color='red')
p=[1,1,1,1,1]
q=[1,2,3,4,5]
plt.scatter(p,q,color='green')
plt.show()
```

```python
x=[10,20,30,40,50]
y=[1,2,3,4,5]
plt.scatter(x,y,color=['red','green','blue','pink','cyan'])
plt.show()
```

```python
import random
x=[10,20,30,40,50]
y=[1,2,3,4,5]
plt.scatter(x,y,c=[random.randint(0,100) for i in range(5)],cmap='viridis')
plt.show()
```

```python
x=[10,20,30,40,50]
y=[1,2,3,4,5]
plt.scatter(x,y,s=[10,20,40,80,160])
plt.show()
```

```python
x=[10,20,30,40,50]
y=[1,2,3,4,5]
plt.scatter(x,y,s=[10,20,40,80,160],alpha=0.9)
plt.show()
```

# **Bars**

```python
categories=['A','B','C','D']
y=[10,20,30,40]
plt.bar(categories,y)
plt.show()
```

```python
categories=['A','B','C','D']
y=[10,20,30,40]
plt.barh(categories,y)
plt.show()
```

```python
categories=['A','B','C','D']
y=[10,20,30,40]
plt.bar(categories,y,color=['cyan','magenta','red','pink'])
plt.show()
```

```python
categories=['A','B','C','D']
y=[10,20,30,40]
plt.bar(categories,y,width=0.5)
plt.show()
```

```python
categories=['A','B','C','D']
y=[10,20,30,40]
plt.barh(categories,y,height=.1)
plt.show()
```

# **piechart**

```python
a=[10,20,30,40]
b=['I','II','III','IV']
c=['red','green','blue','cyan']
plt.pie(a,labels=b,colors=c,startangle=0,explode=[0.2,0,0,0],shadow=True)
plt.show()
```

```python
a=[10,20,30,40]
b=['I','II','III','IV']
c=['red','green','blue','cyan']
plt.pie(a,labels=b,colors=c,startangle=0,explode=[0.2,0,0,0],shadow=True)
plt.legend()
plt.show()
```

```python
a=[10,20,30,40]
b=['I','II','III','IV']
c=['red','green','blue','cyan']
plt.pie(a,labels=b,colors=c,startangle=0,explode=[0.2,0,0,0],shadow=True)
plt.legend(title='4 Categories:-')
plt.show()
```

```python
a=[1,2,3,40]
b=['I','II','III','IV']
c=['red','green','blue','cyan']
plt.pie(a,labels=b,colors=c,startangle=0,autopct='%.1f%%')
plt.show()
```

```python
a=[1,2,3,40]
b=['I','II','III','IV']
c=['red','green','blue','cyan']
plt.pie(a,labels=b,colors=c,startangle=0,autopct='%.1f%%',wedgeprops={'edgecolor':'magenta','linewidth':4})
plt.show()
```
