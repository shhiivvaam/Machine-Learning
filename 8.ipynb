```python
import matplotlib.pyplot as plt
import random
x = [10,20,30,40,50]
y = [1,2,3,4,5]
plt.scatter(x,y,c = [random(0,100)] for i in range(5),cmap='Accent')
plt.show()
```

```python
import matplotlib.pyplot as plt
import random
x = [10,20,30,40,50]
y = [1,2,3,4,5]
plt.scatter(x,y,s = (10,20,30,40,50),cmap='Accent')
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
a = [10,20,30,40]
b = ['I','II','III','IV']
c = ['red','green','blue','cyan']
plt.pie(a,labels=b,colors=c,startangle=0,explode=[.2,0,0,0],shadow=True,autopct='%.1f%%',wedgeprops={'edgecolor':'magenta','linewidth':'4'})
plt.legend()
plt.show()
```
