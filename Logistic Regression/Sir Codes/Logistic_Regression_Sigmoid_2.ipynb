```python
from sklearn.linear_model import LogisticRegression 
model=LogisticRegression()
model.fit(X,y)
tt=model.predict([[2.5]])
print(tt)
```

```python
import numpy as np
xnew=np.linspace(0,3,1000).reshape(-1,1)
y_prob=model.predict_proba(xnew)
print(xnew) 
print('--------------------')
print(y_prob)
```
